```python
# 1. # Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import linprog

# Define constraints
x = np.linspace(0, 10, 200)  # Define x values for plotting

# Constraint equations
y1 = (12 - 2 * x) / 3  # From 2x + 3y <= 12 (machine time)
y2 = (8 - x) / 2       # From x + 2y <= 8 (raw materials)

# Feasible region
y = np.minimum(y1, y2)  # Take the minimum y-values from constraints

# Plot constraints
plt.figure(figsize=(8, 8))
plt.plot(x, y1, label=r'$2x + 3y \leq 12$', color='blue')
plt.plot(x, y2, label=r'$x + 2y \leq 8$', color='orange')

# Shade the feasible region
plt.fill_between(x, 0, y, where=(y >= 0), color='lightblue', alpha=0.5, label='Feasible Region')

# Highlight axes
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Set plot limits
plt.xlim(0, 10)
plt.ylim(0, 10)

# Add labels, title, and legend
plt.xlabel('x (Product A)')
plt.ylabel('y (Product B)')
plt.title('Graphical Solution: Maximizing Profit for a Factory')
plt.legend()
plt.grid()

# Solve using corner points
# Manually calculate intersection points
points = [
    (0, 0),  # Intersection of axes
    (0, 4),  # Intersection of x=0 and y2
    (3, 2),  # Intersection of y1 and y2
    (6, 0)   # Intersection of y1 and x-axis
]

# Evaluate the objective function at corner points
profits = [(3 * x + 4 * y) for x, y in points]
optimal_point = points[np.argmax(profits)]  # Find the point with the maximum profit
optimal_value = max(profits)

# Annotate corner points
for point in points:
    plt.scatter(*point, color='red')
    plt.text(point[0] + 0.2, point[1] + 0.2, f"{point}", color='red')

# Highlight optimal point
plt.scatter(*optimal_point, color='green', s=100, label=f'Optimal Point {optimal_point}')
plt.legend()

# Show plot
plt.show()

# Print results
print(f"Corner Points: {points}")
print(f"Objective Function Values at Corner Points: {profits}")
print(f"Optimal Solution: Point {optimal_point} with Profit = {optimal_value}")
```
